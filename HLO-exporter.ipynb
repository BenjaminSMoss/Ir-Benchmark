```python
def read_spectral_HLO(filename):
    heads, datadict=read_hlo(filename)
    data=pd.DataFrame(datadict)
    head_all=list(heads.get('optional').values())[0]
    head_all.insert(0,'TT')
    data.columns=head_all
    data['TT']=data['TT']-data['TT'][0]
    data.set_index('TT',inplace=True)
    data.drop(data.index[0],inplace=True)
    filename_=end_path1+'.pkl'
    data.to_pickle(filename_)
    print(filename_)

    return data

read_spectral_HLO(filename)




    
```
